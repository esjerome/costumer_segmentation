```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
import seaborn as sns
from operator import attrgetter
import matplotlib.colors as mcolors
```

```python
import numpy as np
```

```python
#!wget https://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx
```

```python
df = pd.read_excel('Online Retail.xlsx',
                   dtype={'CustomerID': str,
                          'InvoiceID': str},
                   parse_dates=['InvoiceDate'])

df.head()
```

```python
df.dropna(subset=['CustomerID'], inplace=True)
```

```python
df.describe().transpose()
```

```python
n_orders = df.groupby(['CustomerID'])['InvoiceNo'].nunique()
mult_orders_perc = np.sum(n_orders > 1) / df['CustomerID'].nunique()
print(f'{100 * mult_orders_perc:.2f}% of customers ordered more than once.')
```

```python
ax = sns.distplot(n_orders, kde=False, hist=True)
ax.set(title='Distribution of number of orders per customer',
       xlabel='# of orders', 
       ylabel='# of customers');
```

```python
df = df[['CustomerID', 'InvoiceNo', 'InvoiceDate','Quantity','UnitPrice']].drop_duplicates()
df['order_month'] = df['InvoiceDate'].dt.to_period('M')
df['cohort'] = df.groupby('CustomerID')['InvoiceDate'] \
                 .transform('min') \
                 .dt.to_period('M') 
```

```python
df_cohort = df.groupby(['cohort', 'order_month']) \
              .agg(n_customers=('CustomerID', 'nunique')) \
              .reset_index(drop=False)
df_cohort['period_number'] = (df_cohort.order_month - df_cohort.cohort).apply(attrgetter('n'))
```

```python
cohort_pivot = df_cohort.pivot_table(index = 'cohort',
                                     columns = 'period_number',
                                     values = 'n_customers')
```

```python
cohort_size = cohort_pivot.iloc[:,0]
retention_matrix = cohort_pivot.divide(cohort_size, axis = 0)
```

```python
with sns.axes_style("white"):
    fig, ax = plt.subplots(1, 2, figsize=(12, 8), sharey=True, gridspec_kw={'width_ratios': [1, 11]})
    
    # retention matrix
    sns.heatmap(retention_matrix, 
                mask=retention_matrix.isnull(), 
                annot=True, 
                fmt='.0%', 
                cmap='RdYlGn', 
                ax=ax[1])
    ax[1].set_title('Monthly Cohorts: User Retention', fontsize=16)
    ax[1].set(xlabel='# of periods',
              ylabel='')

    # cohort size
    cohort_size_df = pd.DataFrame(cohort_size).rename(columns={0: 'cohort_size'})
    white_cmap = mcolors.ListedColormap(['white'])
    sns.heatmap(cohort_size_df, 
                annot=True, 
                cbar=False, 
                fmt='g', 
                cmap=white_cmap, 
                ax=ax[0])

    fig.tight_layout()
```

```python
grouping = df_cohort.merge(df, on=['order_month','cohort'], how='left').groupby(['cohort', 'period_number'])
cohort_data = grouping['Quantity'].mean()
cohort_data = cohort_data.reset_index()
average_price = cohort_data.pivot(index='cohort', columns='period_number', values='Quantity')
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.set_title('Average Quantity by Monthly Cohorts')

sns.heatmap(average_price, annot=True, fmt=".1f", annot_kws={"size": 12}, ax=ax)
b, t = ax.set_ylim() 
b += 0.5 
t -= 0.5 
ax.set_ylim(b, t) 
plt.show()
```

```python
grouping = df_cohort.merge(df, on=['cohort','order_month'], how='left').groupby(['cohort', 'period_number'])
cohort_data = grouping['UnitPrice'].mean()
cohort_data = cohort_data.reset_index()
average_quantity = cohort_data.pivot(index='cohort', columns='period_number', values='UnitPrice')
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.set_title('Average Spend by Monthly Cohorts')

sns.heatmap(average_quantity, annot=True, fmt=".1f", annot_kws={"size": 12}, ax=ax)
b, t = ax.set_ylim() 
b += 0.5 
t -= 0.5 
ax.set_ylim(b, t) 
plt.show()
```
